```python
# Import Libraries

from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
# Set up Splinter

executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Set to the Mars homepage

url = 'https://redplanetscience.com/'
browser.visit(url)
```

```python
# Parse the HTML

html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
html_soup
```

After reviewing the above, it appears that each of the headlines that appears on the main page has the class "content_title" and the blurb of text below it has the class "article_teaser_body".  Both have the HTML tag "div".

```python
# Check to see if the titles can be scraped

titles = html_soup.find_all('div', class_="content_title")

for title in titles:
    heading = title.text
    print(heading)
```

```python
# Check to see if the text blurb can be scraped

blurbs = html_soup.find_all('div', class_="article_teaser_body")

for blurb in blurbs:
    teaser = blurb.text
    print(teaser)
```

```python
# Move the text from the Beautiful Soup objects into python lists

titles_list = []
blurbs_list = []

for title in titles:
    titles_list.append(title.text)

for blurb in blurbs:
    blurbs_list.append(blurb.text)
```

```python
# Create a temporary dictionary for each preview and title and then append my final list
# At the end print the final list of dictionaries

final_list = []

for i in range(len(titles_list)):
    temp_dictionary = {'title': titles_list[i], 'preview': blurbs_list[i]}
    final_list.append(temp_dictionary)
    
print(final_list)
```
